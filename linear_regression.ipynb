```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x  =  np.arange(0,5,1)
y = x
x_cont = np.arange(0,5,0.1)
y_cont = x_cont 
```

```python
plt.figure(figsize=(5,5))
plt.plot(x_cont, y_cont, "b-", linewidth =2)
plt.plot(x, y, "ro", markersize =6)
plt.grid(True)
plt.show()
```

```python

theta1 = 0.5

h = theta1 * x


plt.figure(figsize=(5,5))
plt.plot(x, y, "ro", markersize =6)
plt.plot(x, h, "g--", linewidth = 2)
plt.grid(True)
plt.show()
```

```python
m = x.size
theta1 = np.arange(-3,5,0.1)
h = np.array([theta*x for theta in theta1])

j = ((1/(2*m))*((h-y)**2).sum(axis=1))

plt.figure(figsize=(5,5))
plt.plot(theta1, j, "b-", linewidth = 2)
plt.grid(True)
plt.show()
```

```python
x = np.arange(-2,2,0.1)
y = 2.5 + 5*x
```

```python
m = x.size
theta0 = np.arange(-12.5,17.5,0.1)
theta1 = np.arange(-10,20,0.1)
h = np.array([[(th0+(th1*x)) for th0 in theta0 ]for th1 in theta1])

j = ((1/(2*m))*((h-y)**2).sum(axis=2))
```

```python
fig = plt.figure()
ax = plt.axes(projection="3d")
th0, th1 = np.meshgrid(theta0, theta1)
surf = ax.plot_surface(th0,th1,j, cmap="coolwarm")
```
