```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import numpy as np
import  matplotlib.pyplot as plt
```

```python
x = np.random.rand(200, 1)
```

```python
y =  0.3*x**3 +  25*x**2 + 1.5*np.random.randn(200,1)
```

```python
lr = LinearRegression()
pr = LinearRegression()
```

```python
quadratic = PolynomialFeatures(degree=3)
```

```python
x_quad = quadratic.fit_transform(x)
```

```python
lr.fit(x, y)
```

```python
x_fit = np.arange(0,1, 1/100)

x_fit =x_fit[:, np.newaxis]
```

```python
y_lin_fit = lr.predict(x_fit)
```

```python
pr.fit(x_quad, y)
```

```python
y_quad_fit = pr.predict(quadratic.fit_transform(x_fit))
```

```python
plt.plot(x, y, "bo" )
plt.plot(x_fit, y_lin_fit)
plt.plot(x_fit, y_quad_fit ,"r-")
plt.tight_layout()
plt.show()
```

```python
y_lin_predict = lr.predict(x)
```

```python
y_quad_predict = pr.predict(x_quad)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mean_squared_error(y, y_quad_predict)
```

```python
r2_score(y, y_quad_predict)
```
